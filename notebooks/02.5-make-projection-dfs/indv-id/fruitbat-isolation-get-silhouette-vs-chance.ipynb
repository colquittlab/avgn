### cluster

```python
len(syllable_df)
```

```python
def norm(x):
    return (x-np.min(x)) / (np.max(x) - np.min(x))
```

```python
specs = list(syllable_df.spectrogram.values)
specs = [norm(i) for i in specs]
specs_flattened = flatten_spectrograms(specs)
np.shape(specs_flattened)
```

```python
draw_spec_set(specs, zoom=1, maxrows=10, colsize=25)
```

### cluster

```python
from cuml.manifold.umap import UMAP as cumlUMAP
```

```python
cuml_umap = cumlUMAP()
embedding = cuml_umap.fit_transform(specs_flattened)
```

```python
np.shape(specs_flattened)
```

```python
z  = embedding
```

```python
z = np.vstack(z)
```

```python
fig, ax = plt.subplots(figsize=(15, 15))
#scatter_projections(projection=z, alpha=0.05, labels=None, s=5, ax=ax)
scatter_projections(
    projection=z[syllable_df.indv.values != "0"],
    alpha=1,
    labels=syllable_df.indv.values[syllable_df.indv.values != "0"],
    s=10,
    ax=ax,
)

# ax.set_xlim([-15,15])
```

### get silhouette score

```python
from scipy.stats import kruskal
```

```python
from sklearn.metrics import silhouette_score, silhouette_samples
```

```python
samples = silhouette_samples(list(z), labels = subset_df.indv.values)
chance_samples = silhouette_samples(list(z), labels = np.random.permutation(subset_df.indv.values))
```

```python
kruskal(samples, chance_samples)
```
