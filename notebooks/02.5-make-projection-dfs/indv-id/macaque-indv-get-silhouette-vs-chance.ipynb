### cluster

```python
specs = list(syllable_df.spectrogram.values)
specs = [i / np.max(i) for i in specs]
specs_flattened = flatten_spectrograms(specs)
np.shape(specs_flattened)
```

```python
from cuml.manifold.umap import UMAP as cumlUMAP
```

```python
cuml_umap = cumlUMAP()
z = np.vstack(list(cuml_umap.fit_transform(specs_flattened)))
```

### variation across populations

### get silhouette score

```python
from scipy.stats import kruskal
```

```python
from sklearn.metrics import silhouette_score, silhouette_samples
```

```python
samples = silhouette_samples(list(z), labels = syllable_df.indv.values)
chance_samples = silhouette_samples(list(z), labels = np.random.permutation(syllable_df.indv.values))
```

```python
kruskal(samples, chance_samples)
```

```python
print(np.mean(samples))
```
