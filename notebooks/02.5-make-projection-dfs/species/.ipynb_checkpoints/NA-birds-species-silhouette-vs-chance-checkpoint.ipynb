### cluster

```python
len(syllable_df)
```

```python
specs = list(syllable_df.spectrogram.values)
specs = [i / np.max(i) for i in specs]
specs_flattened = flatten_spectrograms(specs)
np.shape(specs_flattened)
```

```python
from cuml.manifold.umap import UMAP as cumlUMAP

cuml_umap = cumlUMAP()
embedding = cuml_umap.fit_transform(specs_flattened)
```

```python
np.shape(specs_flattened)
```

```python
z  = embedding
```

```python
z = np.vstack(z)
```

```python
fig, ax = plt.subplots(figsize=(15,15))
scatter_projections(projection=z, alpha=1, labels = syllable_df.species.values, s=10, ax = ax, color_palette="tab20")
ax.set_xlim([-15,15])
ax.set_ylim([-15,15])
plt.show()
```

### silhouette

```python
from scipy.stats import kruskal
```

```python
from sklearn.metrics import silhouette_score, silhouette_samples
```

```python
samples = silhouette_samples(list(z), labels = syllable_df.species.values)
chance_samples = silhouette_samples(list(z), labels = np.random.permutation(syllable_df.species.values))
```

```python
kruskal(samples, chance_samples)
```

```python
print(np.mean(samples))
```

```python
from sklearn.metrics import silhouette_score
```

```python
score = silhouette_score(list(z), labels = syllable_df.species.values)
score
```
