### cluster

```python
len(syllable_df)
```

```python
specs = list(syllable_df.spectrogram.values)
specs = [i / np.max(i) for i in specs]
specs_flattened = flatten_spectrograms(specs)
np.shape(specs_flattened)
```

```python
from cuml.manifold.umap import UMAP as cumlUMAP
```

```python
cuml_umap = cumlUMAP()
embedding = cuml_umap.fit_transform(specs_flattened)
```

```python
np.shape(specs_flattened)
```

```python
z = embedding
```

### variation across populations

### get silhouette score for each representation

```python
from sklearn.metrics import silhouette_score
```

```python
from scipy.stats import kruskal
```

```python
from sklearn.metrics import silhouette_score, silhouette_samples
```

```python
silhouette_umap = silhouette_samples(list(z), labels = syllable_df.common_name.values)
chance_silhouette_umap = silhouette_samples(list(z), labels = np.random.permutation(syllable_df.common_name.values))
```

```python
kruskal(silhouette_umap, chance_silhouette_umap)
```

```python
print(np.mean(silhouette_umap))
```

### get silhouette score from spectrum

```python
import scipy.signal

spectrums = np.array([scipy.signal.welch(i)[1] for i in syllable_df.audio.values]);
np.shape(spectrums)
```

```python
from sklearn.metrics import silhouette_score
```

```python
from scipy.stats import kruskal
```

```python
from sklearn.metrics import silhouette_score, silhouette_samples
```

```python
silhouette_spectrum = silhouette_samples(list(spectrums), labels = syllable_df.common_name.values)
chance_silhouette_spectrum = silhouette_samples(list(spectrums), labels = np.random.permutation(syllable_df.common_name.values))
```

```python
kruskal(silhouette_spectrum, chance_silhouette_spectrum)
```

```python
print(np.mean(silhouette_spectrum))
```

```python
plt.plot(spectrums.T, alpha = 0.01);
```

```python
from cuml.manifold.umap import UMAP as cumlUMAP
```

```python
cuml_umap = cumlUMAP()
embedding_spectrum = cuml_umap.fit_transform(spectrums)
```

```python
np.shape(specs_flattened)
```

```python
z = embedding_spectrum
```

### variation across populations

```python
fig, ax = plt.subplots(figsize=(15,15))
scatter_projections(projection=z, alpha=1, labels = syllable_df.common_name.values, s=10, ax = ax, color_palette = 'Set1')
#ax.set_xlim([-15,15])
#ax.set_ylim([-15,15])
plt.show()
```

```python
score = silhouette_score(list(embedding_spectrum), labels = syllable_df.common_name.values)
```

```python
print(score)
```

```python
silhouette_spectrum_umap = silhouette_samples(list(embedding_spectrum), labels = syllable_df.common_name.values)
chance_silhouette_spectrum_umap = silhouette_samples(list(embedding_spectrum), labels = np.random.permutation(syllable_df.common_name.values))
```

```python
kruskal(silhouette_spectrum_umap, chance_silhouette_spectrum_umap)
```

```python
print(np.mean(silhouette_spectrum_umap))
```

##### Compare silhouette of spectrum vs umap/spectrum
