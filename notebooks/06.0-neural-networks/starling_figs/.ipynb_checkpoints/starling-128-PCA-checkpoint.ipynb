```python
from avgn.signalprocessing.create_spectrogram_dataset import flatten_spectrograms
```

```python
# umap
specs_flattened = flatten_spectrograms(all_x)#[:, :10]#[:20000]
```

```python
np.shape(specs_flattened)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=128)
```

```python
pca.fit(specs_flattened)
```

```python
zpca = pca.transform(specs_flattened)
```

```python
proj = pca.inverse_transform(zpca[0])
```

```python
fig, axs = plt.subplots(ncols=2, figsize=(10,5))
axs[0].matshow(specs_flattened[0].reshape((128,128)))
axs[1].matshow(proj.reshape((128,128)))
```

```python
pt1, pt2, pt3 = pca.transform(flatten_spectrograms(np.array([x1*255, x2*255, x3*255]).astype('uint8')))
```

```python
interp_len = 5
```

```python
#pt1x,pt2x,pt3x =gen_func(tf.stack([pt1,pt2,pt3]))

#get proportions
z_list = []
for ci, C in enumerate(np.linspace(0, 1, interp_len)):
    for bi, B in enumerate(np.linspace(0, 1, interp_len)):
        A = 1 - C - B
        z_list.append(
            C * pt1 +
            B * pt2 + 
            A * pt3
        )
z_list = np.vstack(z_list)
# get X
x_list = pca.inverse_transform(z_list)
```

```python
np.shape(x_list)
```

```python
x_list = x_list.reshape(len(x_list), 128, 128, 1)
```

```python


# make diagram
Jdiagram = np.ones((x_list.shape[1]*(interp_len), x_list.shape[2]*(interp_len+2), x_list.shape[3]))
np.shape(Jdiagram)

#populate
i = 0
for ci, C in enumerate(np.linspace(0, 1, interp_len)):
    for bi, B in enumerate(np.linspace(0, 1, interp_len)):
        Jdiagram[(interp_len -1 - bi)*x_list.shape[1]:((interp_len - bi))*x_list.shape[1], (ci+1)*x_list.shape[2]:(ci+2)*x_list.shape[2], :] = x_list[i]*255
        i+=1
Jdiagram[(interp_len - 1)*x_list.shape[1]: (interp_len)*x_list.shape[1], 
        :x_list.shape[2], :] = x3*255

Jdiagram[(interp_len - 1)*x_list.shape[1]: (interp_len)*x_list.shape[1], 
        (interp_len +1)*x_list.shape[2]: (interp_len+2)*x_list.shape[2] , :] = x1*255

Jdiagram[: x_list.shape[1], 
        :x_list.shape[2], :] = x2
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.matshow(np.squeeze(Jdiagram), vmin = 0, cmap=plt.cm.afmhot, origin = 'lower')
ax.axis('off')
save_fig(FIGURE_DIR / 'networks' / 'starling128'/ ('PCA_JDIAGRAM_128'), dpi=300, save_jpg=True)
```
