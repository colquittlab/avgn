### get silhouette score for each representation

```python
from sklearn.metrics import silhouette_score
```

```python
feature_types = [
    ["specs", "Spectrograms"],
    ["PAF", "PAFs"],
    ["UMAP_spec", "UMAP on Spectrograms"],
    ["UMAP_PAF", "UMAP on PAFs"],
    #["LDA_PAF", "LDA on PAF"],
    #["LDA_spec", "LDA on Spectrogram"],
]
```

```python
score_df = pd.DataFrame(columns=["Feature", "Silhouette score", "silhouette-chance"])
for [ft, flabel] in tqdm(feature_types):
    labels = syllable_df["labels"].values
    # labels = labels[syllable_df.train_mask == False]
    feats = np.array([np.array(i).flatten() for i in syllable_df[ft].values])
    # feats = feats[syllable_df.train_mask == False]

    chance_silhouette_score = Parallel(n_jobs=-1)(
        delayed(silhouette_score)(list(feats), np.random.permutation(labels))
        for i in tqdm(range(1000), leave=False)
    )

    score = silhouette_score(list(feats), labels=labels)
    score_df.loc[len(score_df)] = [flabel, score, chance_silhouette_score]
score_df
```

```python
print(score_df.round(3).to_latex(index=False))
```
