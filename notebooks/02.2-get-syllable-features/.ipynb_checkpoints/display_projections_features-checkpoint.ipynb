```python
features = [
    "fund","cvfund", "maxfund","minfund","F1", "sal", # "F3",
    "rms","maxAmp","meanS","stdS","skewS","kurtS","entS",
    "q1","q2","q3","meanT","stdT","skewT","kurtT","entT", "syll_len"]
len(features)
```

```python
import matplotlib.image as mpimg
```

```python
ncols = 4
nrows =6
zoom = 8
fig, axs = plt.subplots(nrows = 6, ncols=ncols, figsize=(zoom*ncols,nrows*zoom), gridspec_kw = {'wspace':0, 'hspace':0.0})
for fi, feature in tqdm(enumerate(features)):
    ax = axs.flatten()[fi]
    img=mpimg.imread((FIGURE_DIR / 'cassins-specs-vs-feats' / 'feats_noborder' / (feature + '.jpg')).as_posix())
    ax.imshow(img)
    ax.axis('off')
    ax.set_title(feature, fontsize=34)
    
for fi in range(fi, nrows*ncols):
    ax = axs.flatten()[fi]
    ax.axis('off')

ensure_dir(FIGURE_DIR / 'cassins-specs-vs-feats')
save_fig(FIGURE_DIR / 'cassins-specs-vs-feats' / 'umap_feats_proj', dpi=300, save_pdf=True)
plt.show()
```
