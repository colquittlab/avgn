```python
DATASET_ID = 'swamp_sparrow'
hparams = HParams()
# create a dataset object
dataset = DataSet(DATASET_ID, hparams = hparams)
```

```python
population = [
    str(i)
    for i in np.array(
        [
            dataset.data_files[key].data["POPID"] if key in dataset.data_files else None
            for key in tqdm(syllable_df.key.values)
        ]
    )
]
np.unique(population)
```

```python
population = np.array(population)
```

```python
pd.DataFrame({i:[np.sum(population == i)] for i in np.unique(population)}).T.sort_values(by=0, ascending = False)
```

```python
clusterer = hdbscan.HDBSCAN(
    min_cluster_size=int(len(umap_embedding) * 0.01), # the smallest size we would expect a cluster to be
    min_samples=1, # larger values = more conservative clustering
)
clusterer.fit(umap_embedding);
#clusterer.labels_
```

```python
nex = -1
scatter_spec(
    umap_embedding[:nex],
    specs[:nex],
    column_size=8,
    #x_range = [-4.5,4],
    #y_range = [-4.5,5.5],
    pal_color="hls",
    color_points=False,
    enlarge_points=0,
    figsize=(10, 10),
    range_pad = 0.1,
    scatter_kwargs = {
        'labels': clusterer.labels_[:nex],
        'alpha':1,
        's': 0.25,
        'show_legend': True,
        'log_x':False,
        'color_palette': 'tab20',
    },
    matshow_kwargs = {
        'cmap': plt.cm.Greys
    },
    line_kwargs = {
        'lw':3,
        'ls':"dashed",
        'alpha':0.25,
    },
    draw_lines=True,
    n_subset= 1000,
    border_line_width = 3,

);
#save_fig(FIGURE_DIR / 'discrete_umap' / 'swamp-sparrow-taper', dpi=300, save_jpg=True, save_png=True)
```

#### plot spectrograms from categories

```python
from avgn.visualization.spectrogram import draw_spec_set
```

```python
unique_clusters = np.unique(clusterer.labels_)[1:]
len(unique_clusters)
```

```python
zm = 6
ncols=2
nrows=6
fig, axs = plt.subplots(nrows=nrows, ncols =ncols, figsize= (ncols*zm, nrows*zm))
for li, label in enumerate(unique_clusters):
    ax = axs.flatten()[li]
    draw_spec_set(np.array(specs)[clusterer.labels_ == label], zoom=1, maxrows=5, colsize=5, ax=ax)
    ax.set_title(label, fontsize=24)
for li in range(li, len(unique_clusters)+1):
    ax = axs.flatten()[li]
    ax.axis('off')
plt.show();
```
